Принято решение изучать количество задач total_sales. Готовим данные, сглаживаем ряд окном в 7 дней (обусловлено наличием сезонности - больше продаж в выходные дни).

```python
df['date'] = pd.to_datetime(df['date'])
df_time_series = df.groupby('date')['total_sales'].sum()
df_smoothed = df_time_series.rolling(window=7).mean()
df_smoothed = df_smoothed.dropna()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df_time_series, label="Исходные данные", alpha=0.5, linestyle="-", color="blue")
plt.plot(df_smoothed, label="Сглаженный ряд (окно 7 дней)", linestyle="solid", color="red", linewidth=2)
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())
plt.xticks(rotation=45)
plt.title("Total Sales по дням (до и после сглаживания)")
plt.xlabel("Дата")
plt.ylabel("Total Sales")
plt.legend()
plt.grid()
plt.show()
```

Проверяем оба ряда (сглаженный и нет) на стационарность с помощью расширенного теста Дики-Фуллера

```python
def test_stationarity(timeseries):
    result = adfuller(timeseries.dropna())
    print("ADF Test:")
    print(f"ADF Statistic: {result[0]:.4f}")
    print(f"p-value: {result[1]:.4f}")
    if result[1] <= 0.05:
        print("Ряд стационарен")
        return True
    else:
        print("Ряд нестационарен")
        return False
```

```python
is_stationary = test_stationarity(df_time_series)
if not is_stationary:
    df_stationary = df_time_series.diff().dropna()
    is_stationary_after = test_stationarity(df_stationary)
```

```python
is_stationary_smoothed = test_stationarity(df_smoothed)
if not is_stationary_smoothed:
    df_stationary_smoothed = df_smoothed.diff().dropna()
    is_stationary_after_smoothed = test_stationarity(df_stationary)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df_stationary, label="Стационарный ряд (после преобразований)", color="green")
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())
plt.xticks(rotation=45)
plt.title("Стационарный временной ряд")
plt.xlabel("Дата")
plt.ylabel("Изменения Total Sales")
plt.legend()
plt.grid()
plt.show()
```

Строим графики автокорреляции и частичной автокорреляции. Ряд без сглаживания показывает сезонность каждые 7 дней. Это подтверждает правильность выбора окна сглаживания.

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 5))
plot_acf(df_smoothed.dropna(), ax=ax[0], lags=30)
plot_pacf(df_smoothed.dropna(), ax=ax[1], lags=50)
ax[0].set_title("Автокорреляционная для сглаженных данных (ACF)")
ax[1].set_title("Частичная автокорреляционная для сглаженных данных (PACF)")
plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 5))
plot_acf(df_stationary.dropna(), ax=ax[0], lags=50)
plot_pacf(df_stationary.dropna(), ax=ax[1], lags=50)
ax[0].set_title("Автокорреляционная функция (ACF)")
ax[1].set_title("Частичная автокорреляционная функция (PACF)")
plt.show()
```

Подбираем лучшие параметры и строим модель SARIMAX 

```python

p = range(0, 3)  
d = range(0, 2)  
q = range(0, 3) 
P = range(0, 2)  
D = range(0, 2)  
Q = range(0, 2)  
s = [7]  

aic_values = []
bic_values = []
hqic_values = []
param_combinations = []

for param in itertools.product(p, d, q):
    for seasonal_param in itertools.product(P, D, Q, s):
        try:
            model = SARIMAX(df_stationary.dropna(),
                            order=param,
                            seasonal_order=seasonal_param,
                            enforce_stationarity=False,
                            enforce_invertibility=False)
            result = model.fit(disp=0)

            aic_values.append(result.aic)
            bic_values.append(result.bic)
            hqic_values.append(result.hqic)
            param_combinations.append((param, seasonal_param))

        except Exception as e:
            print(f"Ошибка для {param} x {seasonal_param}: {e}")

```

```python
min_aic_index = np.argmin(aic_values)

best_params = param_combinations[min_aic_index]
best_aic = aic_values[min_aic_index]

print(f"Лучшие параметры: {best_params}")
print(f"Минимальный AIC: {best_aic}")

max_aic_index = np.argmax(aic_values)
mean_aic = np.mean(aic_values)

worst_params = param_combinations[max_aic_index]
worst_aic = aic_values[max_aic_index]

print(f"Худшие параметры (максимальный AIC): {worst_params}")
print(f"Максимальный AIC: {worst_aic}")

print(f"Среднее значение AIC: {mean_aic}")
```

```python
best_p, best_d, best_q = best_params[0]
best_P, best_D, best_Q, best_s = best_params[1]

model_sarima = SARIMAX(df_stationary,
                       order=(best_p, best_d, best_q),
                       seasonal_order=(best_P, best_D, best_Q, best_s),
                       enforce_stationarity=False,
                       enforce_invertibility=False)

sarima_result = model_sarima.fit()
print(sarima_result.summary())
```

```python
future_steps = 10
forecast = sarima_result.forecast(steps=future_steps)
forecast_index = pd.date_range(df_stationary.index[-1], periods=future_steps + 1, freq='D')[1:] 
```

```python
plt.plot(df_stationary, label="Исторические данные")
plt.plot(forecast_index, forecast, label="Прогноз", color="red")
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())
plt.xticks(rotation=45)
plt.xlabel("Дата")
plt.ylabel("Продажи")
plt.legend()
plt.grid()
plt.show()
```

```python
ps = range(0, 3)  
ds = range(0, 2)  
qs = range(0, 3) 
Ps = range(0, 2)  
Ds = range(0, 2)  
Qs = range(0, 2)  
ss = [7]  

best_aic_s = float("inf")
best_params_s = None
for param in itertools.product(p, d, q):
    for seasonal_param in itertools.product(P, D, Q, s):
        try:
            model = SARIMAX(df_stationary.dropna(), 
                            order=param,
                            seasonal_order=seasonal_param,
                            enforce_stationarity=False,
                            enforce_invertibility=False)
            result = model.fit(disp=0) 
            if result.aic < best_aic_s:
                best_aic_s = result.aic
                best_params_s = (param, seasonal_param)
        except Exception as e:
            print(f"Ошибка для {param} x {seasonal_param}: {e}")  

print(f"\nЛучшие параметры: ARIMA{best_params_s[0]} x SARIMA{best_params_s[1]} с AIC={best_aic_s}")


```

```python
best_ps, best_ds, best_qs = best_params_s[0]
best_Ps, best_Ds, best_Qs, best_ss = best_params_s[1]

model_sarima_s = SARIMAX(df_smoothed,
                       order=(best_ps, best_ds, best_qs),
                       seasonal_order=(best_Ps, best_Ds, best_Qs, best_ss),
                       enforce_stationarity=False,
                       enforce_invertibility=False)

sarima_result_s = model_sarima_s.fit()
print(sarima_result_s.summary())
```

```python
future_steps = 10
forecast_s = sarima_result_s.forecast(steps=future_steps)
forecast_index_s = pd.date_range(df_smoothed.index[-1], periods=future_steps + 1, freq='D')[1:] 
```

```python
plt.plot(df_smoothed, label="Исторические данные")
plt.plot(forecast_index_s, forecast_s, label="Прогноз", color="red")
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())
plt.xticks(rotation=45)
plt.xlabel("Дата")
plt.ylabel("Продажи")
plt.legend()
plt.grid()
plt.show()
```

Модель смогла учесть сезонность на ряде без сглаживания и уловить направление на сглаженном ряде. 

Теперь строим прогноз с помощью сети долгой краткосрочной памяти 

```python
scaler = MinMaxScaler(feature_range=(0, 1))
df_scaled = scaler.fit_transform((df_stationary.values.reshape(-1, 1)))
```

```python
look_back = 10
```

```python
def create_sequences(data, look_back):
    X, y = [], []
    for i in range(look_back, len(data)):
        X.append(data[i - look_back:i, 0])
        y.append(data[i, 0])
    return np.array(X), np.array(y)

X, y = create_sequences(df_scaled, look_back)
```

```python
split = int(len(X) * 0.8)
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]
```

```python
X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))
X_test = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))
```

```python
model = Sequential()
model.add(LSTM(50, activation='relu', input_shape=(look_back, 1)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')

model.fit(X_train, y_train, epochs=15, batch_size=32, verbose=1)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred_inv = scaler.inverse_transform(y_pred)
y_test_inv = scaler.inverse_transform(y_test.reshape(-1, 1))
full_series = scaler.inverse_transform(df_scaled)
```

```python
test_index_start = look_back + split
test_dates = df_stationary.index[test_index_start:test_index_start + len(y_test)]
```

```python
plt.figure(figsize=(14, 6))
plt.plot(df_stationary.index, full_series, label='Весь временной ряд')
plt.plot(test_dates, y_pred_inv, color='red', label='Прогноз (LSTM)')
plt.title('Прогноз временного ряда (LSTM)')
plt.legend()
plt.grid(True)
plt.show()
```
